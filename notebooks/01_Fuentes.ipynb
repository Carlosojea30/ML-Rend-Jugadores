### Creación del target 

```python
# Orden temporal por jugador

df = df.sort_values(by=["player_name", "season"])
df.head()
```

```python
# Convertir temporada a año inicial (ej: "2016-17" -> 2016)

df["season_start"] = df["season"].str[:4].astype(int)
```

```python
# Target: puntos por partido en la siguiente temporada

df["pts_next_season"] = df.groupby("player_name")["pts"].shift(-1)
```

```python
# Quitamos filas sin siguiente temporada (no se puede predecir)

df = df.dropna(subset=["pts_next_season"])
```
